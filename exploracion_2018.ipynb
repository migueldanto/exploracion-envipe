### Probando los datos que vienen en la tabla tpervic1, tpervic2
saber cual es subconjunto de otras

```python
merg1=tsdem[["ID_PER","EDAD","SEXO"]].merge(tper_vic1[["ID_PER","EDAD","SEXO"]],left_index=False,right_index=False,on="ID_PER",how="inner")
print(merg1.shape)
merg1.head()
```

```python
merg2=merg1.merge(tper_vic1[["ID_PER","EDAD","SEXO"]],left_index=False,right_index=False,on="ID_PER",how="inner")
print(merg2.shape)
merg2.head()
```

**lo anterior indica indica que** la tabla tpervic1 y tepervic2 son un subconjunto de la tabla tsdem, es decir, 
que que de las 326 mil personas que contiene la tabla tsdem que podria interpretarse como el catalogo de personas consultadas,
solo 91541 personas contestaron la encuesta de persepcion 1 y la encuesta de percepcion2 

- y todas las que contestaron la de percepcion 1 son exactamente las mismas que contestaron la de percepcion2

### revisando los datos de tmod_vic
Se supone que son los delitos que ya les sucedieron a las personas encuestadas, haciendo un conteo agrupado por personas, 
se podra saber cuantas y cuales personas son las que estan contempladas en esta tabla

```python
group_11=tmod_vic[["ID_PER","ID_DEL"]].groupby(by="ID_PER").count()
print(group_11.shape[0],"Es el numero de personas registradas en esta tabla")
group_11.head()
```

```python
merg3=merg2.merge(group_11,left_index=False,right_index=True,left_on="ID_PER",how="inner")
print(merg3.shape)
merg3.head()
```

**Lo anterior dice que** de las 91 541 personas que contestaron percepcion 1 y percepcion 2, hay 26572 personas que contestaron 
haber sufrido algun delito
