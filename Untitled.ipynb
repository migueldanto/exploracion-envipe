# Estimando el numero de delitos

```python
import pandas as pd
codes,camp1,camp2=[],[],[]
for x in range(1,5):
    delito_code=str(x).zfill(2)
    campo_exist="AP6_4_"+delito_code
    campo_conteo="AP6_6_"+delito_code
    codes.append(delito_code)
    camp1.append(campo_exist)
    camp2.append(campo_conteo)
    print(delito_code,campo_exist,campo_conteo)
    #agregar a una lista
```

```python
for x in range(5,16):
    delito_code=str(x).zfill(2)
    campo_exist="AP7_3_"+delito_code
    campo_conteo="AP7_4_"+delito_code
    print(delito_code,campo_exist,campo_conteo)
    codes.append(delito_code)
    camp1.append(campo_exist)
    camp2.append(campo_conteo)
```

```python
campos_conteo=pd.DataFrame({"code":codes,"existe":camp1,"cantidad":camp2})
campos_conteo
```

```python
#cargando los datos
from simpledbf import Dbf5
base_path="datos_2018/bd_envipe2018_dbf/BASE_DE_DATOS_ENVIPE_2018_dbf/"
dbf_tper_vic2=Dbf5(base_path+"TPer_Vic2.dbf")
tper_vic2=dbf_tper_vic2.to_dataframe()
print(tper_vic2.shape)
tper_vic2.head(3)
```

```python
prb_existe=campos_conteo.loc[0]["existe"]
prb_cantidad=campos_conteo.loc[0]["cantidad"]
tper_vic2[tper_vic2[prb_existe]=="1"][[prb_cantidad,prb_existe]].sort_values(by="AP6_6_01",ascending=False).head(4)
```

```python
def calcula_canti(row,ex,ca):
    return float(row[ca]) if row[ex]=="1" else 0
encuestas_cop=tper_vic2.copy()
for i,row in campos_conteo.iterrows():
    cm_exs=row["existe"]
    cm_cant=row["cantidad"]
    encuestas_cop[row["code"]+"_count" ]=encuestas_cop.apply(calcula_canti,args=(cm_exs,cm_cant),axis=1)
    
```

```python
import numpy as np
campos_ver=np.array([ [row["existe"],row["cantidad"],row["code"]+"_count" ] for i,row in campos_conteo.iterrows()]).flatten()
campos_ver=np.insert(campos_ver,0,"ID_PER")
encuestas_cop[campos_ver.tolist()].head()
```

```python
campos_to_count=[row["code"]+"_count"  for i,row in campos_conteo.iterrows() ]
encuestas_cop["sumas_d"]=encuestas_cop[campos_to_count].sum(axis=1, skipna=True)
encuestas_cop[["ID_PER"]+campos_to_count+["sumas_d"]].head()
```

```python
encuestas_cop["sumas_d"].sum()
```

```python
#el conteo de delitos!!!
encuestas_cop[campos_to_count].sum(axis=0, skipna=True)
```
